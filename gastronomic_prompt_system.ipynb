## Introducción

Los restaurantes dependen cada vez más de redes sociales como Instagram para atraer clientes y comunicar su identidad gastronómica. Sin embargo, generar contenido atractivo de forma constante requiere tiempo, creatividad y una estrategia clara.

La inteligencia artificial permite agilizar este proceso, pero la calidad del contenido generado depende en gran medida del prompt utilizado. Cuando las instrucciones son vagas, los resultados suelen ser genéricos y poco útiles.

Este proyecto propone desarrollar un sistema de prompts optimizados para generar contenido gastronómico utilizando inteligencia artificial.

## Objetivo

El objetivo de esta prueba de concepto es demostrar cómo la optimización de prompts mejora la generación de contenido gastronómico con inteligencia artificial.

Para ello se compararán prompts simples con prompts optimizados para analizar cómo el contexto, el tono y las instrucciones influyen en la calidad del resultado.

## Metodología

En este notebook se comparan prompts simples y prompts optimizados para distintos tipos de contenido gastronómico.

Se analizarán cuatro casos:

1. Caption para Instagram
2. Ideas de contenido gastronómico
3. Descripción de un plato
4. Prompt para generación de imagen

En cada caso se comparará un prompt básico con un prompt optimizado que incluye contexto, rol, objetivo y formato.

## Caso 1: Caption para Instagram

En este caso se busca generar un caption para Instagram sobre un restaurante que sirve pasta fresca.

Primero se utiliza un prompt simple y luego un prompt optimizado para observar la diferencia.

```python
prompt_simple = """
Escribe un caption para Instagram sobre un restaurante que sirve pasta fresca.
"""

print(prompt_simple)
```

```python
prompt_optimizado = """
Actúa como un creador de contenido gastronómico especializado en restaurantes italianos.

Escribe un caption para Instagram sobre un restaurante que sirve pasta fresca hecha a mano todos los días.

El objetivo es transmitir la experiencia de comer en el lugar, no solo describir el plato.

El tono debe ser sensorial, elegante y atractivo.

Evita frases genéricas o clichés.

Escribe entre 100 y 150 palabras en un solo párrafo.
"""

print(prompt_optimizado)
```

### Análisis

El prompt simple ofrece una instrucción muy general, por lo que el resultado puede ser genérico.

El prompt optimizado incorpora contexto, tono, objetivo y formato, lo que permite generar un contenido más específico y alineado con la experiencia gastronómica.

## Caso 2: Ideas de contenido gastronómico

En este caso se generan ideas de contenido para redes sociales de un restaurante.

```python
prompt_simple_ideas = """
Dame 5 ideas de contenido para Instagram para un restaurante.
"""

print(prompt_simple_ideas)
```

```python
prompt_ideas = """
Actúa como un estratega de contenido especializado en restaurantes.

Genera 5 ideas de contenido para Instagram para un restaurante contemporáneo.

Las ideas deben ayudar a construir identidad de marca.

Incluye ideas relacionadas con:
- el proceso de cocina
- la atmósfera del restaurante
- el equipo
- los detalles gastronómicos

Evita ideas genéricas.
"""

print(prompt_ideas)
```

### Análisis

El prompt optimizado orienta al modelo hacia una estrategia de contenido más rica y variada, en lugar de limitarse a mostrar platos.

## Caso 3: Descripción de un plato

```python
prompt_plato = """
Actúa como un escritor gastronómico.

Describe un plato de pasta fresca servido en un restaurante elegante.

El objetivo es transmitir textura, aroma y experiencia gastronómica.

Evita frases genéricas.

Escribe entre 80 y 120 palabras.
"""

print(prompt_plato)
```

## Caso 4: Prompt para generación de imagen gastronómica

```python
prompt_imagen = """
Genera una imagen realista de un plato de pasta fresca servido en un restaurante elegante.

La iluminación debe ser cálida.
El foco debe estar en la textura de la pasta.
El fondo debe ser simple y sofisticado.
La escena debe transmitir una experiencia gastronómica atractiva.
"""

print(prompt_imagen)
```

## Técnicas de Fast Prompting aplicadas

En este proyecto se aplicaron las siguientes técnicas:

- asignación de rol
- inclusión de contexto
- definición de objetivo
- especificación del tono
- definición del formato de salida

Estas técnicas permiten guiar mejor a la inteligencia artificial y obtener resultados más útiles.

## Conclusión

La prueba de concepto demuestra que la calidad del contenido generado por inteligencia artificial depende en gran medida de la construcción del prompt.

Los prompts optimizados permiten obtener resultados más específicos, creativos y alineados con la identidad de un restaurante.

Esto demuestra que el diseño estratégico de prompts puede ser una herramienta valiosa para la creación de contenido gastronómico en redes sociales.
